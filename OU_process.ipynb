# Ornstein-Uhlenbeck process

Standalone demonstration of the OU process used as the reference noise model in the thesis. 

```python
import numpy as np
import matplotlib.pyplot as plt

theta = 0.1        # mean-reversion rate
mu = 0             # long-run mean
sigma = 0.3        # noise amplitude
dt = 0.01          # output sampling step
T = 100            # total simulated time (s)
```

```python
N = int(T/dt)
x = np.zeros(N)
x[0] = 2           # start away from the mean to show reversion

for t in range(1, N):
    x[t] = x[t-1] + theta*(mu - x[t-1])*dt + sigma*np.sqrt(dt)*np.random.randn()

time = np.linspace(0, T, N)
plt.plot(time, x)
plt.xlabel("time")
plt.ylabel("x")
plt.title("OU Process")
plt.show()
```

```python
# Five independent realisations from the same parameters
for i in range(5):
    x = np.zeros(N)
    for t in range(1, N):
        x[t] = x[t-1] + theta*(mu-x[t-1])*dt + sigma*np.sqrt(dt)*np.random.randn()
    plt.plot(time, x, alpha=0.7)

plt.show()
```
